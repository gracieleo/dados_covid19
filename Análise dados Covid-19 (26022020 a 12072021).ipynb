```python
import requests as r
```

```python
final_data = []                      #criar uma nova lista com alguns dos parametros
for obs in raw_data:
    final_data.append([obs['Confirmed'], obs['Deaths'],obs['Recovered'],obs['Active'],obs['Date']])
```

```python
final_data.insert(0,['confirmados', 'mortes','recuperados','ativos','data'])    #lista de listas add o header na posição 0
final_data
```

```python
confirmados = 0                                    #referenciar posição na lista (de acordo com o header)
mortes = 1
recuperados = 2
ativos = 3
data = 4
```

```python
for i in range(1, len(final_data)):
    final_data[i][data] = final_data[i][data][:10]      
```

```python
def get_datasets(y,labels):                           #função cria dados
    if type(y[0] == list):                            #se tipo = lista
        datasets = []                                 #inicia lista vazia
        for i in range(len(y)):                       #para cada posição dentro do comprimento de y
            datasets.append({                         #add 
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return[
            {
                'label':labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title =''):       #função define titulo
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return{
        'title':title,
        'display': display
        }
        
```

```python
def create_chart(x, y, labels, kind='bar',title=''):       #função cria grafico
    datasets = get_datasets(y,labels)
    options = set_title(title)
    
    chart = {
        'type':kind,
        'data':{
            'labels': x,
            'datasets':datasets
        },
        'options':options
    }
   
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')             #retorna arquivo de imagem
    return resp.content
```

```python
def save_image(path, content):                     #salva imagem
    with open(path,'wb') as image:                 #wb escrita binario
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display 
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for obs in final_data[1::10]:                           #pula de 10 em 10 dias funções slice
    y_data_1.append(obs[confirmados])
    
    
y_data_2 = []
for obs in final_data[1::10]: 
    y_data_2.append(obs[recuperados])
    
labels = ['confirmados','recuperados']

x = []
for obs in final_data[1::10]:
    x.append(obs[data].strftime('%d/%m/%Y'))  #transforma data em string

chart = create_chart(x,[y_data_1,y_data_2], labels, title = 'Gráfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-gráfico.png',chart_content)
display_image('meu-primeiro-gráfico.png')
```

```python
from urllib.parse import quote
    
```

```python
def get_api_qrcode(link):
    text = quote(link)                      #parsing do link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
